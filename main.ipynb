### Нормализация

```python
import re
from razdel import sentenize

def normalize(text):
    if not text:
        return ""
    text = text.replace("\xa0", " ")
    text = re.sub(r"\s+", " ", text) 
    return text.strip()

def split_sentences(text):
    if not text:
        return []
    
    sentences = []
    for s in sentenize(text):
        # clean_s = normalize(s.text).lower()  # 
        clean_s = normalize(s.text)
        if clean_s:
            sentences.append(clean_s)
            
    return sentences

print(data[0]["text"])
sentences = split_sentences(data[0]["text"])
print(f"Количество предложений: {len(sentences)}")
print(f"Первые 3 предложения:")
for j, s in enumerate(sentences[:3]):
    print(f"{j}: {s}")
```

### Лемматизация

```python
import pymorphy2
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

morph = pymorphy2.MorphAnalyzer()
stop_words = set(stopwords.words('russian'))
print(stop_words)
```

```python
def lemmatize_sentence(sent):
    words = re.findall(r"\w+", sent.lower())
    lemmas = []
    for w in words:
        if w not in stop_words:
            lemmas.append(morph.parse(w)[0].normal_form)
    return " ".join(lemmas)

def lemmatize_text(sentences):
    return [lemmatize_sentence(s) for s in sentences]
```

```python
test = data[0]["text"]
test_lemma = lemmatize_text(sentences)

print(f"До: {test}")
print(f"После: {test_lemma}")
print(f"Длина после лемматизации: {len(test_lemma)}")
```

### Векторизация

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def build_tfidf(sentences_lemmas):
    vectorizer = TfidfVectorizer(
        max_df=0.95,
        min_df=1,
        max_features=20000,
        ngram_range=(1, 1),
        tokenizer=str.split,
        preprocessor=None,
        token_pattern=None,
        lowercase=False
    )
    X = vectorizer.fit_transform(sentences_lemmas)
    return X, vectorizer

def add_position_bias(scores, coef=0.3):
    n = len(scores)
    pos = np.linspace(1, 0.2, n)
    return scores + coef * pos

def mmr_order(sents, base, sim, alpha=0.7, limit=300, min_k=1):
    sel, rest = [], list(range(len(sents)))
    while rest:
        best_i, best_score = None, -1e9
        for i in rest:
            sim_to_sel = max(sim[i, sel]) if sel else 0
            score = alpha * base[i] - (1 - alpha) * sim_to_sel
            if score > best_score:
                best_score, best_i = score, i
        new_sel = sel + [best_i]
        summary = " ".join(sents[j] for j in sorted(new_sel))
        if len(summary) > limit and len(new_sel) > min_k:
            break
        sel = new_sel
        rest.remove(best_i)
    return sel
```

### Суммаризация

```python
def summarize(text, limit=300, use_mmr=False, d=0.85, position_coef=0.18, min_sent_len=30, alpha=0.7):
    sents = split_sentences(text)
    if not sents:
        return ""
    
    sents = [s for s in sents if len(s) >= min_sent_len]
    if not sents:
        return ""
    
    if sum(len(s) for s in sents) <= limit:
        return " ".join(sents)

    sents_lem = [lemmatize_sentence(s) for s in sents]
    X, _ = build_tfidf(sents_lem)
    sim = cosine_similarity(X)

    np.fill_diagonal(sim, 0)
    sim[sim < 0.1] = 0

    row = sim.sum(1, keepdims=True)
    row[row == 0] = 1
    M = sim / row

    pr = np.ones(len(sents)) / len(sents)
    for _ in range(50):
        pr_new = (1 - d) / len(sents) + d * M.T.dot(pr)
        if np.linalg.norm(pr_new - pr, 1) < 1e-6: 
            break
        pr = pr_new
    
    scores = pr / np.log1p([len(s) for s in sents])
    scores = add_position_bias(scores, coef=position_coef) 
    
    if use_mmr:
        selected = mmr_order(sents, scores, sim, alpha=alpha, limit=limit)
        summary = " ".join(sents[i] for i in sorted(selected))
    else:
        idx_sorted = np.argsort(scores)[::-1]
        used = set()
        total = 0
        
        for i in idx_sorted:
            cand_len = len(sents[i]) + 1  
            if total + cand_len <= limit:
                used.add(i)
                total += cand_len
            if total >= limit * 0.9:
                break
        
        chosen = [sents[i] for i in sorted(used)]
        summary = " ".join(chosen)
        
        # Контроль лимита
        while len(summary) > limit and chosen:
            chosen.pop()
            summary = " ".join(chosen)
    
    return summary
```

### Проверка и оценка результатов через ROUGE

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=False)

def evaluate_model(data, summarize_function, show_examples=20):
    all_rouge1 = []
    all_rouge2 = []
    all_rougeL = []

    for i in range(500):
        ref = data[i]["summary"]
        pred = summarize_function(data[i]["text"])
        
        scores = scorer.score(ref, pred)
        all_rouge1.append(scores['rouge1'].fmeasure)
        all_rouge2.append(scores['rouge2'].fmeasure)
        all_rougeL.append(scores['rougeL'].fmeasure)

        if i <= show_examples:
            print(f"\nСтатья {i}: R1={scores['rouge1'].fmeasure:.3f} | R2={scores['rouge2'].fmeasure:.3f} | RL={scores['rougeL'].fmeasure:.3f}")
            print("Реферат:", ref)
            print("Длина реферата:", len(ref), "символов")
            print("Реферат модели:", pred)
            print("Длина реферата:", len(pred), "символов")
    
    return all_rouge1, all_rouge2, all_rougeL

def table_rouge(rouge1, rouge2, rougeL):
    print(f"\n{'Метрика':<12} {'Среднее':<10} {'Медиана':<10} {'0':<6} {'(0;0.5)':<8} {'[0.5;1)':<8} {'1':<4}")
    print("-" * 65)

    for name, scores in [("ROUGE-1", rouge1), ("ROUGE-2", rouge2), ("ROUGE-L", rougeL)]:
        mean_val = np.mean(scores)
        median_val = np.median(scores)
        c0 = sum(1 for r in scores if r == 0)
        c1 = sum(1 for r in scores if 0 < r < 0.5)
        c2 = sum(1 for r in scores if 0.5 <= r < 1)
        c3 = sum(1 for r in scores if r == 1)
        print(f"{name:<12} {mean_val:<10.4f} {median_val:<10.4f} {c0:<6} {c1:<8} {c2:<8} {c3:<4}")
```

```python
all_rouge1, all_rouge2, all_rougeL = evaluate_model(data, summarize_function=lambda text: summarize(text, use_mmr=False))
```

```python
table_rouge(all_rouge1, all_rouge2, all_rougeL)
```

### Проверка и оценка работы с оптимизацией MMR

```python
all_rouge1_mmr, all_rouge2_mmr, all_rougeL_mmr = evaluate_model(data, summarize_function=lambda text: summarize(text, use_mmr=True))
```

### Проверка через другие метрики

```python
from bert_score import score

def evaluate_bertscore(references, predictions, lang="ru"):
    model_name = "bert-base-multilingual-cased"
    
    P, R, F1 = score(
        predictions,
        references,
        lang=lang,
        model_type=model_name,
        verbose=False
    )
    
    return P.mean().item(), R.mean().item(), F1.mean().item()

val_ds = load_dataset("IlyaGusev/gazeta", split="validation[:100]")

refs = []
preds = []
for i in range(len(val_ds)):
    refs.append(val_ds[i]["summary"])
    preds.append(summarize(val_ds[i]["text"], limit=300))

p, r, f1 = evaluate_bertscore(refs, preds)
```

```python
print(f"BERTScore: P={p:.4f} | R={r:.4f} | F1={f1:.4f}")
```
